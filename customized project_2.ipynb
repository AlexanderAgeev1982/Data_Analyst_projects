```python
def recovery(C, F, T): # –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è  
    q = (C*(F-T))
    w =(F*(C-T))
    r = q / w * 100
    return r
```

```python
t = mae_recovery['rougher.output.tail_au'] 
f = mae_recovery['rougher.input.feed_au']
c = mae_recovery['rougher.output.concentrate_au']
r = recovery(c, f, t)
print('MAE :', mean_absolute_error(mae_recovery['rougher.output.recovery'], r))
```

## –ú–æ–¥–µ–ª—å

4.1 –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

```python
def smape (target, predictions):
    q = abs(predictions - target)
    w = (abs(target) + abs(predictions)) / 2
    total = (q / w).sum()
    smape = (1 / len(target)) * total * 100
    return smape
```

4.5 –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏.

```python
def cross_val (model, features, target):
    scores = cross_val_score(model, features, 
                                   target, scoring=make_scorer(smape, greater_is_better=True), cv=5,n_jobs=-1)
    smape_result = np.mean(scores)
    return smape_result
```

<h3> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v1</h3>

–ù–µ–º–Ω–æ–≥–æ –ø—Ä–æ –∫–∞—Å—Ç–æ–º–Ω—ã–π —Å–∫–æ—Ä–∏–Ω–≥ –∏ `greater_is_better=False`: –≤–æ—Ç –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–∞—è —Å—Ç–∞—Ç—å—è -¬†https://kiwidamien.github.io/custom-loss-vs-custom-scoring.html
    
–ï—Å–ª–∏ –∫—Ä–∞—Ç–∫–æ, —Ç–æ —Ç–≤–æ–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ cross_val_score —Å–æ —Å–≤–æ–µ–π –º–µ—Ç—Ä–∏–∫–æ–π –∏ —Ñ–ª–∞–≥–æ–º `greater_is_better=False` –æ—Å—Ç–∞–µ—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ, —Ç–æ–ª—å–∫–æ —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, —Ç–∞–∫ –∫–∞–∫ –µ—Å—Ç—å –º–µ—Ç—Ä–∏–∫–∞ (–ø–æ –∫–æ—Ç–æ—Ä–æ–π –º—ã –≤—ã–±–∏—Ä–∞–µ–º –º–æ–¥–µ–ª—å) –∏ –µ—Å—Ç—å –ª–æ—Å—Å (–ø–æ –∫–æ—Ç–æ—Ä–æ–π –æ–±—É—á–∞–µ—Ç—Å—è –º–æ–¥–µ–ª—å). –¢–∞–∫ –≤–æ—Ç make_scorer —Å–æ–∑–¥–∞–µ—Ç –º–µ—Ç—Ä–∏–∫—É –∏–∑ –ª–æ—Å—Å–∞, –ø–æ –∫–æ—Ç–æ—Ä–æ–π —Ç—ã –≤—ã–±–∏—Ä–∞–µ—à—å –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, –Ω–æ —Å–∞–º–∞ –º–æ–¥–µ–ª—å –±—É–¥–µ—Ç –æ–±—É—á–∞—Ç—å—Å—è –ø–æ –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å (—Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ MSE), —Ç–∞–∫ —á—Ç–æ —ç—Ç–æ –Ω—É–∂–Ω–æ —Ç–æ–ª—å–∫–æ –¥–ª—è –≤—ã–±–æ—Ä–∞ –º–æ–¥–µ–ª–∏. –ü—Ä–æ—Å—Ç–æ –ø—Ä–∏–Ω—è—Ç–æ —Å—Ç–∞–≤–∏—Ç—å —Ñ–ª–∞–≥ greater_is_better=False –¥–ª—è –º–µ—Ç—Ä–∏–∫, –∫–æ—Ç–æ—Ä—ã–µ —è–≤–ª—è—é—Ç—Å—è –ª–æ—Å—Å–∞–º–∏ (–Ω–µ –≤—Å–µ –º–µ—Ç—Ä–∏–∫–∏ –ª–æ—Å—Å—ã, –Ω–∞–ø—Ä–∏–º–µ—Ä, recall, precision –∏ —Ç–ø).
    
–ù–û –µ—Å–ª–∏ —Ç—ã –∏—Å–ø–æ–ª—å–∑—É–µ—à—å—Å—è –≥—Ä–∏–¥—Å–µ—Ä—á, —Ç–æ `greater_is_better=False` –Ω—É–∂–Ω–æ —É–∫–∞–∑—ã–≤–∞—Ç—å –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ, —Ç–∞–∫ –∫–∞–∫ best_estimator_ –≤—ã–±–∏—Ä–∞–µ—Ç –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –ø–æ –º–µ—Ç—Ä–∏–∫–µ —É–∂–µ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ =)


```python
# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã–≤–æ–¥–∞ sMAPE –∫–∞–∂–¥–æ–≥–æ —ç—Ç–∞–ø–∞ –∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–π sMAPE.

def ml(model):
    smape_rougher = cross_val (model, rougher_features_train, rougher_target_train)
    smape_final = cross_val (model, final_features_train, final_target_train)
    smape_end = 0.25 * smape_rougher + 0.75 * smape_final
    print('sMAPE –Ω–∞ —ç—Ç–∞–ø–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏ —Å–æ—Å—Ç–∞–≤–∏–ª–∞', smape_rougher)
    print('sMAPE –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º —ç—Ç–∞–ø–µ —Å–æ—Å—Ç–∞–≤–∏–ª–∞', smape_final)
    print('–ö–æ–Ω–µ—á–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ sMAPE —Å–æ—Å—Ç–∞–≤–∏–ª–∞', smape_end)
```

4.7 –°–¥–µ–ª–∞–µ–º –ø–µ—Ä–µ–±–æ—Ä –≥–º–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
best_smape = 10
est_est = 0
best_depth = 0
for est in range(10, 51, 10):
    for depth in range (1, 5):
        model = RandomForestRegressor(n_estimators=est, max_depth=depth)
        smape_rougher = cross_val (model, rougher_features_train, rougher_target_train)
        smape_final = cross_val (model, final_features_train, final_target_train)
        smape_end = 0.25 * smape_rougher + 0.75 * smape_final
        if smape_end < best_smape:
            best_smape = smape_end
            best_est = est
            best_depth = depth
print('–ù–∞–∏–ª—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å sMAPE {}, –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å \
—Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ä–µ–≤—å–µ–≤ {} –∏ –≥–ª—É–±–∏–Ω–æ–π {}'.format(best_smape, best_est, best_depth))
```
